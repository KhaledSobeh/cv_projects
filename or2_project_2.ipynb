the coding solution for the problem, please pass W,w,p,q Only!

```python
def knapsack(W,w,p,q,k = -1,tables = {}):
    if(k==-1): #Starting Point
        k = len(w)+1
        tables[k] = np.zeros(W+1,dtype=int).reshape(W+1,1)
        return knapsack(W,w,p,q,k-1,tables) 
    if(k==0): #End
        return tables
    # k>=1
    x_values = np.arange(int(W/w[k-1])+1) #Legal x values for each container.
    if k==1: #Actually, this will be the last board, so we start with W capacity.
        s_values = [W]
    else:
        s_values = np.arange(W+1) 
    current_table = np.zeros(len(s_values)*len(x_values),dtype=int).reshape(len(s_values),len(x_values)) 
    #This is a 2D array.

    for s in s_values: 
        for x in x_values:
            if len(s_values)==1: #for the last board, we will start with W capacity.
                if x==0:
                    current_table[0,x] = np.amax(tables[k+1],1)[s_values[0]]
                elif x*w[k-1] <= s:
                    current_table[0,x] = x*p[k-1] + np.amax(tables[k+1],1)[s-x*w[k-1]]
                else:
                    current_table[0,x] = -1
            else:
                if x==0:
                    current_table[s,x] = np.amax(tables[k+1],1)[s]
                elif x*w[k-1] <= s:
                    current_table[s,x] = x*p[k-1]*(1-q) + (x-1)*p[k-1]*q + np.amax(tables[k+1],1)[s-x*w[k-1]]
                else:
                    current_table[s,x] = -1

    tables[k] = current_table
    return knapsack(W,w,p,q,k-1,tables)  
```

A function for printing the solution

```python
def print_optimal_tables(tables,W,p):
    space = 0 #f_tables
    optimal_table = {} #f_tables
    opt_full_tables = {}
    for elem in range(1,len(tables.keys())):
        df = pd.DataFrame(tables[elem])
        colname_f = 'f*'+str(elem)
        df[colname_f] = np.amax(df,1)
        colname_x = 'x*'+str(elem)
        df[colname_x] = np.argmax(tables[elem],1)
        if(elem == 1): 
            df.index = [str(W)]
            optimal_table[elem] = np.array(df.iloc[0,-2:])
        if(elem !=1):
            space = space + optimal_table[elem-1][1]*p[elem-2]
            optimal_table[elem] = np.array(df.iloc[W-space,-2:])
        df = df.astype(str)
        df.replace({"-1": "-"}, inplace=True)
        opt_full_tables[elem] = df
    
    for elem in reversed(range(1,len(tables.keys()))):
        if elem == len(tables)-1:
            print('---------------------------------------')
            print('------------- Full tables -------------')
        print('---------------------------------------')
        print('k='+str(elem))
        print(opt_full_tables[elem])
    
    opt_df = pd.DataFrame(optimal_table).transpose()
    opt_df.columns = ['f*','x*']
    print('---------------------------------------')
    print('------------ Optimal table ------------')
    print('---------------------------------------')
    print(opt_df)
```

a function for getting the second ID

```python
def second_id():
    id_lst = []
    id_lst.append(random.choice([2,3])) #first number in id
    help_lst = [i for i in range(1,11)] #0 already replaced with 10
    for i in range(0,8):
        id_lst.append(random.choice(help_lst))
    return(id_lst)
```

Getting the data

```python
w1 = second_id()
w2 = second_id()
w= w1 + w2
p= w2 + w1
W = sum(w)
q = abs(w1[3]-w2[3])/10
```
